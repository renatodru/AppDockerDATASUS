```python
import leitor
import matplotlib as mpl
import matplotlib.pyplot as plt
df_RD = leitor.carregador()
```

```python
df_graf = df_RD[(df_RD['MUNIC_RES']=='410690')][(df_RD['COMPLEX']=='03')][(df_RD['FINANC']=='06')]
categ_sum = df_graf.groupby('DESCR_ESTABELECIMENTO')['VAL_TOT'].sum()
teto = str(round(categ_sum.max()*1.1))
teto = int(str(int(teto[:1])+1)+len(teto[1:])*'0')

plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')
plt.gca().set(  ylim=(0, teto),
                xlabel='Estabelecimentos',
                ylabel='Valor')
plt.xticks(fontsize=12); plt.yticks(fontsize=12)
plt.title("""
Valor total em milhes de R$ das contas hospitalares de procedimentos de alta complexidade
de moradores de curitiba agrupados por estabelecimentos Contratados pelo SUS""", fontsize=18)

categ_sum.plot.bar()
plt.show()
```
